```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3.5 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np

# RNG with seed = 9044
rng = np.random.default_rng(9044)

# Generate sensor readings
sensor_a = rng.normal(25, 3.5, 200)
sensor_b = rng.normal(27, 4.5, 200)

# Generate timestamps from 0 to 10 seconds
timestamps = rng.uniform(0, 10, 200)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.


import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(timestamps, sensor_a, color='tab:blue', alpha=0.7, label='Sensor A')
plt.scatter(timestamps, sensor_b, color='tab:orange', alpha=0.7, label='Sensor B')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor A and Sensor B Temperature Readings vs Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

plt.title("Box Plot of Sensor Temperature Readings")
plt.ylabel("Temperature (°C)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
