```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3.5 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np

# RNG with seed = 9044
rng = np.random.default_rng(9044)

# Generate sensor readings
sensor_a = rng.normal(25, 3.5, 200)
sensor_b = rng.normal(27, 4.5, 200)

# Generate timestamps from 0 to 10 seconds
timestamps = rng.uniform(0, 10, 200)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.


import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(timestamps, sensor_a, color='tab:blue', alpha=0.7, label='Sensor A')
plt.scatter(timestamps, sensor_b, color='tab:orange', alpha=0.7, label='Sensor B')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor A and Sensor B Temperature Readings vs Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(sensor_a, bins=15, color='tab:blue', alpha=0.7, edgecolor='black')

plt.title("Distribution of Sensor A Temperature Readings")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

plt.title("Box Plot of Sensor Temperature Readings")
plt.ylabel("Temperature (°C)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Analysis

Scatter Plot
The scatter plot shows how the temperature readings change over time for both sensors. Sensor A staysmore consistent, while Sensor B has more variation. This matches the code because both sensors were genrated with different means and standard deviations.

Histogram
The histogram shows the distribution of the temperature values. Sensor A's values cluster around its mean, while Sensor B's values are centered higher and spread out more. This reflects the parameters used when generating the random data.

Box Plot
The box plot compares the two sensors directly. Sensor B has a higher median and a larger IQR, meaning it varies more than Sensor A. The whiskers and any outliers also show that Sensor B has a wider range of values.

Overall
All three plots confirm that Sensor A is more stable and Sensor B is hotter and more variable, which matches how the data was created in the code.
